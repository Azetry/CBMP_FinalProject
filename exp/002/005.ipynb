```python
features = {
    'pvalue': list(pd.read_csv("../../data/selected_features/p-value-recur-related-train.csv").columns),
    'corr': list(pd.read_csv("../../data/selected_features/corr-recur-related-train.csv").columns),
    'rf': list(pd.read_csv("../../data/selected_features/rf-recur-related-train.csv").columns)
}
```

```python
for name, feature in features.items():
    print(f"{name}: {len(feature)}")
```

```python
result_all = set(features['pvalue']) & set(features['corr']) & set(features['rf'])
len(result_all)
```

```python
result_corr_rf = set(features['corr']) & set(features['rf'])
len(result_corr_rf)
```

```python
result_corr_pvalue = set(features['corr']) & set(features['pvalue'])
len(result_corr_pvalue)
```

```python
result_pvalue_rf = set(features['pvalue']) & set(features['rf'])
len(result_pvalue_rf)
```

可以從特徵交集了解：
- p-value, correlation 特徵交集比較多
- rf 特徵組與另外兩個特徵組分布不同很多
