Statistic test

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
all_df = pd.read_csv("../../data/raw/tcga_stad_all.csv")
```

```python
all_df
```

```python
recurrence_df = all_df[ all_df.recurrence != -1 ]
metastasis_df = all_df[ all_df.metastasis != -1 ]
```

```python
display(recurrence_df.shape)
display(recurrence_df)
```

#### Correlation

```python
corrmat = recurrence_df.drop(columns=['survival_status', 'survival_time', 'metastasis']).corr()
```

```python
corrmat.nlargest(10, 'recurrence')['recurrence'].index
```

```python
cor_recur_abs = corrmat['recurrence'].apply(np.abs)
recur_related_features = cor_recur_abs[cor_recur_abs > cor_recur_abs.quantile(0.75)].index.values
```

```python
with open("../../data/selected_features/corr-recur-related.csv", "w") as f:
    for idx, feature in enumerate(recur_related_features):
        if idx < len(recur_related_features)-1: f.write(f"{feature},")
        else: f.write(feature)
```

#### Statistic Test 
- samples > 50 : Kolmogorov-Smirnov test (pvalue < 0.05)
- if True(p-value <0.05): Mann-Whitney
- if False: independent T test

```python
features = list(recurrence_df.drop(columns=['survival_status', 'survival_time', 'recurrence', 'metastasis']).columns)
```

```python
# statistics, pvalue = stats.shapiro(recurrence_df['ENSG00000082929.8'])
norm_test = list()
for feature in features:
    statistics, pvalue = stats.kstest(recurrence_df[feature], "norm")
    norm_test.append(pvalue>=0.05)

statistics_df = pd.DataFrame(data={'feature_name':features, 'norm_test': norm_test})
```

```python
display(statistics_df.head())
display(statistics_df.norm_test.value_counts())
```

```python
recur_related_features = list()
for feature in features:
    value_0 = recurrence_df[recurrence_df.recurrence == 1][feature].values
    value_1 = recurrence_df[recurrence_df.recurrence == 0][feature].values
    statistics, pvalue = stats.mannwhitneyu(value_0, value_1)
    if pvalue<0.05: recur_related_features.append(feature)
```

```python
with open("../../data/selected_features/p-value-recur-related.csv", "w") as f:
    for idx, feature in enumerate(recur_related_features):
        if idx < len(recur_related_features)-1: f.write(f"{feature},")
        else: f.write(feature)
```
