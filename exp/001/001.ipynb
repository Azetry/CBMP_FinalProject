```python
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
all_df = pd.read_excel("../../data/raw/tcga_stad.xlsx",sheet_name="all")
```

```python
raw_df = pd.read_excel("../../data/raw/tcga_stad.xlsx",sheet_name="raw")
```

```python
raw_df.rename(columns={"gene_id":"Patient ID"}, inplace=True)
raw_df.set_index("Patient ID", inplace=True)
```

```python
display(all_df)
display(raw_df)
```

```python
display(raw_df['Overall Survival Status'].value_counts())
display(raw_df['recurrence'].value_counts())
```

```python
recurrence_map = {
    '0:DiseaseFree': 0,
    '1:Recurred/Progressed': 1,
    np.NaN: -1,
}
survival_map = {
    '0:LIVING': 0,
    '1:DECEASED': 1
}
metastasis_map = {
    'M0': 0,
    'M1': 1,
    'MX':-1
}
```

```python
clinical_df = pd.read_excel("../../data/raw/stad_tcga_clinical_data.xlsx", sheet_name="工作表1")
clinical_df.set_index("Patient ID", inplace=True)
clinical_df = clinical_df.filter(items=raw_df.index.values, axis=0)
clinical_df['Overall Survival Status'] = clinical_df['Overall Survival Status'].map(survival_map)
clinical_df['Disease Free Status'] = clinical_df['Disease Free Status'].map(recurrence_map)
clinical_df['ajcc_pathologic_m'] = clinical_df['ajcc_pathologic_m'].map(metastasis_map)
clinical_df.rename(columns={'ajcc_pathologic_m':"metastasis"}, inplace=True)
```

```python
all_df['metastasis'] = clinical_df['metastasis'].values
```

```python
all_df.rename(columns={'Overall Survival Status':"survival_status", 'Overall Survival (Months)':"survival_time"}).to_csv("../../data/raw/tcga_stad_all.csv", index=False)
```

```python
display(all_df[all_df.recurrence != -1]['recurrence'].value_counts())
display(all_df[all_df.metastasis != -1]['metastasis'].value_counts())
```
